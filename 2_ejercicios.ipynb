### Utiliza Python y Numpy para realizar las siguientes operaciones:

```python
import numpy as np
```

```python
# ejercicio 1. Definir el arreglo
a = np.array([1, 2, 3, 4])

# 2. Determinar si ninguno de los elementos es cero.
# np.all(): Comprueba si TODOS los resultados de la condición son True.
resultado = np.all(a != 0)

print("Arreglo a:", a)
print("¿Ningún elemento es cero? (Resultado de np.all(a != 0)):", resultado)
```

```python
# ejercicio 1 
b= a>0
b
```

```python
# ejercicio 2

# 1. Definir el arreglo
a = np.array([1, 0, 2, 0])

# 2. Usar np.any() para determinar si algún elemento NO es cero
resultado = np.any(a != 0)

print("Arreglo a:", a)
print("¿Contiene algún elemento que no sea cero?", resultado)
```

```python
# ejercicio 3

# 1. Definir el arreglo
a = np.array([1, 2, np.nan, np.inf])

# 2. Crear las condiciones booleanas
es_infinito = np.isinf(a) 
es_nan = np.isnan(a) 

# 3. Combinar las condiciones con OR (|) y verificar si ALGUNO es True con np.any()
resultado = np.any(es_infinito | es_nan)

print("Arreglo a:", a)
print("¿Contiene valores infinitos o NaN?:", resultado)
```

```python
# ejercicio 4

# 1. Definir el arreglo
a = np.array([1, 2, np.inf])

resultado = np.any(np.isinf(a))

print("Arreglo a:", a)
print("¿Contiene valores infinitos? (Resultado de np.any(np.isinf(a))):", resultado)
```

```python
# ejercicio 5

a = np.array([1, np.nan, 3])

resultado = np.any(np.isnan(a))

print("Arreglo a:", a)
print("¿Contiene valores NaN? (Resultado de np.any(np.isnan(a))):", resultado)
```

```python
# ejercicio 6 

a = np.array([1, 3, 5])
b = np.array([2, 3, 4])

print("Arreglo a:", a)
print("Arreglo b:", b)
print("-" * 10)

# 2. Realizar las comparaciones
# a > b (Mayor que)
mayor_que = a > b
print("a > b (Mayor que):        ", mayor_que)

# a >= b (Mayor o igual que)
mayor_o_igual_que = a >= b
print("a >= b (Mayor o igual que): ", mayor_o_igual_que)

# a < b (Menor que)
menor_que = a < b
print("a < b (Menor que):        ", menor_que)

# a <= b (Menor o igual que)
menor_o_igual_que = a <= b
print("a <= b (Menor o igual que): ", menor_o_igual_que)
```

```python

# ejercicio 7 

# 1. Crear el arreglo
a = np.array([1, 7, 13, 105])

memoria_total = a.itemsize * a.size

print("Arreglo a:", a)
print("Tipo de dato (dtype):", a.dtype)
print("Tamaño de cada elemento (bytes):", a.itemsize)
print("Número total de elementos:", a.size)
print("-" * 40)
print(f"Memoria total ocupada por el arreglo: {memoria_total} bytes")
```

```python
# ejercicio 8 

a = np.arange(30, 71)

print("Arreglo resultante:", a)
print("Forma del arreglo (cantidad de elementos):", a.shape)
```

```python
# ejercicio 9 

# Empezar en 31 (el número después de 30)
# Terminar en 70 (el número antes de 70)
a = np.arange(31, 70)

print("Arreglo resultante (excluyendo 30 y 70):", a)
print("Primer elemento:", a[0])
print("Último elemento:", a[-1]) 
```

```python
# ejercicio 10

a_pares = np.arange(32, 70, 2)

print("Arreglo de números pares entre 30 y 70:", a_pares)
print("Primer elemento:", a_pares[0])   # 32
print("Último elemento:", a_pares[-1])  # 68
```

```python
# ejercicio 14

a = np.arange(15, 56)

print("Arreglo completo (15 a 55):")
print(a)
print("-" * 60)

resultado_ = a[1:-1]

print("Valores intermedios (excluyendo 15 y 55):")
print(resultado_)
```

```python
# ejercicio 15

a = np.linspace(5, 50, 10)

print("Arreglo de 10 valores equitativamente distribuidos:")
print(a)
print("\nForma del arreglo (longitud):", a.shape)
```

```python
# ejercicio 16

vector = np.arange(0, 21)
print("Vector original (0 a 20):")
print(vector)
print("-" * 60)

condicion = (vector >= 9) & (vector <= 15)

vector[condicion] = vector[condicion] * -1

print("Vector con signo cambiado (9 a 15):")
print(vector)
```

```python
# ejercicio 17 

a = np.array([1, 2, 3])
print("Arreglo original (Tipo):", type(a))

lista_python = a.tolist()

print("Lista de Python (Resultado):", lista_python)
print("Tipo del resultado:", type(lista_python))
```

```python
# ejercicio 18 

# 1. Definir la lista nativa de Python
l = [1, 2, 3]
print("Objeto original (Tipo):", type(l))

# 2. Convertir la lista a un arreglo de NumPy usando np.array()
arreglo_numpy = np.array(l)

print("Arreglo de NumPy (Resultado):", arreglo_numpy)
print("Tipo del resultado:", type(arreglo_numpy))
```

```python
# ejercicio 19
#  1. Definir la matriz original
m = np.array([[1, 2], [3, 4]])
print("Matriz original m:")
print(m)
print("-" * 30)

# 2. Verificar la paridad usando el operador módulo (%)
# La condición (m % 2 == 0) devuelve True para todos los números que son divisibles por 2 (pares).
matriz_booleana = (m % 2 == 0)

# 3. Almacenar los resultados en otra matriz (la matriz booleana)
# El resultado es una nueva matriz booleana con las mismas dimensiones que m.
print("Matriz de resultados (True = Par, False = Impar):")
print(matriz_booleana)
```

```python
# ejercicio 20 
# 1. Definir la matriz
m = np.array([[1, 2, 3], [4, 5, 6]])

# 2. Encontrar la forma de la matriz (filas, columnas) usando .shape
filas, columnas = m.shape

print("Matriz m:")
print(m)
print("-" * 30)
print(f"Forma de la matriz (m.shape): {m.shape}")
print(f"Número de filas: {filas}")
print(f"Número de columnas: {columnas}")
```
